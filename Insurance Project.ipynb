```python
#Select input and output from given dataset before apply sampling tech
X=df.drop("insuranceclaim",axis=1)
Y=df["insuranceclaim"]
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=1,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()

X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)

print("After Scaling the training data")
print(X_train)
print("After Scaling the testing data")
print(X_test)
```

```python
pd.Series(Y_train).value_counts()
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
def create_model(model):
  model.fit(X_train,Y_train) #we train the model
  Y_pred=model.predict(X_test) #test
  print(classification_report(Y_test,Y_pred))
  print("Confusion Matrix :")
  #Confusion Matrix
  print(confusion_matrix(Y_test,Y_pred))
  return model
```

```python
# 1. forward selection
```

```python
X=df.drop('insuranceclaim',axis=1)
Y=df['insuranceclaim']
```

```python
columns=[]
for col in X:
  columns.append(col)
  X_new=df[columns]
  X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
  lr=LogisticRegression()
  lr.fit(X_train,Y_train)
  score=lr.score(X_test,Y_test)
  print('Column: ',col,'Score: ',score)
```

```python
# 2. Backward Selection
```

```python
columns=[]
for col in X:
  columns.append(col)
  X_new=df[columns]
  X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
  lr=LogisticRegression()
  lr.fit(X_train,Y_train)
  score=lr.score(X_test,Y_test)
  print('Column: ',col,'Score: ',score)
  X_new=X_new.iloc[:,:-1]
  print('After Removing column', col)
```
